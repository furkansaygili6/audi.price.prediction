```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("auidi.csv")
```

```python
df.head(3)
```

```python
df=df.drop(columns=["href","index","PPY","MileageRank", "PriceRank","PPYRank","Score"])
```

```python
df.head(3)
```

```python
df.info()
```

```python
df["Engine"]= df["Engine"].str.replace("L","")
```

```python
df["Engine"]=pd.to_numeric(df["Engine"])
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=["Type", "Transmission", "Fuel"], drop_first=True)
```

```python
df.head(5)
```

```python
y=df["Price(£)"]
x=df.drop(columns=["Price(£)"])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=70)
```

```python
li=LinearRegression()
model=li.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```
